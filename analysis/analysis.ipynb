# Analysis of the UjiIndoorLoc dataset

## Imports

```python
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
```

## Read dataset

```python
# Simple configuration
TRAINING_DATA_PATH = '/Users/timhager/Library/Mobile Documents/com~apple~CloudDocs/Masterstudium/Auslandssemester/INAR/project/dataset/TrainingData.csv'
VALIDATION_DATA_PATH = '/Users/timhager/Library/Mobile Documents/com~apple~CloudDocs/Masterstudium/Auslandssemester/INAR/project/dataset/ValidationData.csv'
```

```python
training_data = pd.read_csv(TRAINING_DATA_PATH)
validation_data = pd.read_csv(VALIDATION_DATA_PATH)

# Remove unnecessary columns
drop_columns = ["RELATIVEPOSITION", "USERID", "PHONEID", "TIMESTAMP"]
training_data = training_data.drop(columns=drop_columns)
validation_data = validation_data.drop(columns=drop_columns)
```

## First analysis for better overview

```python
# Shapes
print("Training data shape:", training_data.shape)
print("Validation data shape:", validation_data.shape)

print()

# Heads
print("Training data:")
print(training_data.head())
print("\nValidation data:")
print(validation_data.head())

print()

# Info
print("Training data info:")
print(training_data.info())
print("\nValidation data info:")
print(validation_data.info())

print()

# Describe
print("Training data description:")
print(training_data.describe())
print("\nValidation data description:")
print(validation_data.describe())

print()

# Check for NaN and Null
print("Training data NaN values:")
print(training_data.isna().sum())
print("\nValidation data NaN values:")
print(validation_data.isna().sum())

print()

# Check for duplicates
print("Training data duplicate rows:", training_data.duplicated().sum())
print("Validation data duplicate rows:", validation_data.duplicated().sum())
```
